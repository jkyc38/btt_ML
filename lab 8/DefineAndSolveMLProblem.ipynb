```python
new_names = {'Life Ladder': 'Happiness', 
            'Social support': 'Support', 
            'Healthy life expectancy at birth': 'Life Expectancy', 
            'Freedom to make life choices': 'Freedom', 
            'Perceptions of corruption': 'Corruption', 
            'GINI index (World Bank estimate)': 'GINI index'
            }

df = df.rename(columns=new_names)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['Democratic Quality', 'Delivery Quality',
                      'gini of household income reported in Gallup, by wp5-year',
                      'GINI index (World Bank estimate), average 2000-15',
                     'country', 'year'])
```

```python
df.isnull().sum()
```

```python
confidence_gov_mean = df['Confidence in national government'].mean()
gini_mean = df['GINI index'].mean()

df['Confidence in national government'].fillna(value=confidence_gov_mean, inplace=True)
df['GINI index'].fillna(value=gini_mean, inplace=True)

```

```python
df.isnull().sum()
```

```python
null_values = ['Log GDP per capita', 'Life Expectancy', 'Support', 'Freedom', 'Generosity', 'Corruption', 'Positive affect', 'Negative affect']
for col in null_values:
    col_mean = df[col].mean()
    df[col].fillna(value=col_mean, inplace=True)

df.isnull().sum()
#feature engineering step finished
```

```python
df.head()
```

```python
X = df.drop(columns='Life Expectancy')
y= df['Life Expectancy']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
X_train
```

```python
y_train.shape
```

```python
#init linear regression
linear_model = LinearRegression()
linear_model.fit(X_train,y_train)
y_pred = linear_model.predict(X_test)


# The mean squared error
print('\nModel Performance\n\nRMSE =   %.2f'
      % np.sqrt(mean_squared_error(y_test, y_pred)))
# The coefficient of determination: 1 is perfect prediction
print(' R^2 =   %.2f'
      % r2_score(y_test, y_pred))


```
